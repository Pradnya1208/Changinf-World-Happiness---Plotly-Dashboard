```python
data_2015 = pd.read_csv("2015.csv")
data_2016 = pd.read_csv("2016.csv")
data_2017 = pd.read_csv("2017.csv")
data_2018 = pd.read_csv("2018.csv")
data_2019 = pd.read_csv("2019.csv")
data_2020 = pd.read_csv("2020.csv")
```

```python
data_2015["upperwhisker"] = data_2015["Happiness Score"] + 2*data_2015["Standard Error"]
data_2015["lowerwhisker"] = data_2015["Happiness Score"] - 2*data_2015["Standard Error"]
data_2015["Year"] = 2015
```

```python
d_2015 = data_2015[["Country", "Region", "Year", "Happiness Rank", "Happiness Score", "upperwhisker", 
                    "lowerwhisker", "Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Freedom",
                    "Generosity", "Trust (Government Corruption)", "Dystopia Residual"]]
d_2015.rename(columns = {'Family':'Family/Social Status', 'Freedom':'freedom to make life choices', 
                              'Trust (Government Corruption)':'Perceptions of corruption'}, inplace = True)
d_2015
```

```python
data_2016["Year"] = 2016
d_2016 = data_2016[["Country", "Region", "Year", "Happiness Rank", "Happiness Score",  "Upper Confidence Interval","Lower Confidence Interval",
          "Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)", 
          "Generosity", "Dystopia Residual"]]
d_2016.rename(columns = {'Upper Confidence Interval':'upperwhisker', 'Lower Confidence Interval':'lowerwhisker', 'Family':'Family/Social Status', 'Freedom':'freedom to make life choices', 
                              'Trust (Government Corruption)':'Perceptions of corruption'}, inplace = True)
d_2016
```

```python
Region_2017 =[]
country_2017 = []
for i in data_2017["Country"].values:
    if i in data_2016["Country"].values:
        Region_2017.append(data_2016[data_2016.Country == i]["Region"].values[0])
        country_2017.append(i)
    else:
        if i == "Lesotho":
            Region_2017.append("Sub-Saharan Africa")
        if i == "Central African Republic":
            Region_2017.append("Sub-Saharan Africa")
        if i == "Hong Kong S.A.R., China":
            Region_2017.append("Eastern Asia")
        if i == "Taiwan Province of China":
            Region_2017.append("Eastern Asia")
        if i == "Mozambique":
            Region_2017.append("Sub-Saharan Africa")
   
            
def Diff(li1, li2):
    return (list(list(set(li1)-set(li2)) + list(set(li2)-set(li1))))


#print(Diff(data_2017["Country"].values, country_2017))
data_2017["Year"] = 2017
data_2017["Region"] = Region_2017
d_2017 = data_2017[["Country", "Region", "Year", "Happiness.Rank", "Happiness.Score", "Whisker.high", "Whisker.low", "Economy..GDP.per.Capita.",
           "Family", "Health..Life.Expectancy.", "Freedom", "Generosity", "Trust..Government.Corruption.",
           "Dystopia.Residual"]]

d_2017.rename(columns = {'Whisker.high':'upperwhisker', 'Whisker.low':'lowerwhisker', 'Family':'Family/Social Status', 'Freedom':'freedom to make life choices', 
                              'Trust..Government.Corruption.':'Perceptions of corruption', 'Happiness.Rank': 'Happiness Rank', 
                        'Happiness.Score': 'Happiness Score', 'Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 
                        'Health..Life.Expectancy.': 'Health (Life Expectancy)', 'Dystopia.Residual': 'Dystopia Residual'}, inplace = True)
d_2017
```

```python
Region_2018 =[]
country_2018 = []
for i in data_2018["Country or region"].values:
    if i in data_2016["Country"].values:
        Region_2018.append(data_2016[data_2016.Country == i]["Region"].values[0])
        country_2018.append(i)
    else:
        if i == "Lesotho":
            Region_2018.append("Sub-Saharan Africa")
        if i == "Central African Republic":
            Region_2018.append("Sub-Saharan Africa")
        if i == "Northern Cyprus":
            Region_2018.append("Western Europe")
        if i == "Trinidad & Tobago":
            Region_2018.append("Latin America and Caribbean")
        if i == "Mozambique":
            Region_2018.append("Sub-Saharan Africa")




#print(Diff(data_2018["Country or region"].values, country_2018))
data_2018["Year"] = 2018
data_2018["Region"] = Region_2018
data_2018["upperwhisker"] = "No data"
data_2018["lowerwhisker"] = "No data"

data_2018["Dystopia_residual"] = "No Data"

d_2018 = data_2018[["Country or region", "Region","Year", "Overall rank", "Score", "upperwhisker", "lowerwhisker", "GDP per capita", "Social support", 
                    "Healthy life expectancy","Freedom to make life choices", "Generosity", "Perceptions of corruption",
                    "Dystopia_residual"]]

d_2018.rename(columns = {'Country or region':'Country', 'Social support':'Family/Social Status', 'Freedom to make life choices':'freedom to make life choices', 
                         'Overall rank': 'Happiness Rank', 
                         'Score': 'Happiness Score', 
                         'GDP per capita':'Economy (GDP per Capita)', 
                        'Healthy life expectancy': 'Health (Life Expectancy)', 'Dystopia_residual': 'Dystopia Residual'}, inplace = True)
d_2018
```

```python
Region_2019 =[]
country_2019 = []
for i in data_2019["Country or region"].values:
    if i in data_2016["Country"].values:
        Region_2019.append(data_2016[data_2016.Country == i]["Region"].values[0])
        country_2019.append(i)
    else:
        if i == "Lesotho":
            Region_2019.append("Sub-Saharan Africa")
        if i == "Central African Republic":
            Region_2019.append("Sub-Saharan Africa")
        if i == "Northern Cyprus":
            Region_2019.append("Western Europe")
        if i == "Trinidad & Tobago":
            Region_2019.append("Latin America and Caribbean")
        if i == "Mozambique":
            Region_2019.append("Sub-Saharan Africa")
        if i == "North Macedonia":
            Region_2019.append("Central and Eastern Europe")
        if i == "Gambia":
            Region_2019.append("Sub-Saharan Africa")
        if i == "Swaziland":
            Region_2019.append("Sub-Saharan Africa")

#print(Diff(data_2019["Country or region"].values, country_2019))
            
data_2019["Year"] = 2019
data_2019["upperwhisker"] = "No data"
data_2019["lowerwhisker"] = "No data"
data_2019["Region"] = Region_2019
data_2019["Dystopia_residual"] = "No Data"
#Add region, empty whiskers
d_2019 = data_2019[["Country or region", "Region", "Year", "Overall rank", "Score", "upperwhisker", "lowerwhisker", 
                    "GDP per capita", "Social support", "Healthy life expectancy",
          "Freedom to make life choices", "Generosity", "Perceptions of corruption", "Dystopia_residual"]]


d_2019.rename(columns = {'Country or region':'Country', 'Social support':'Family/Social Status', 'Freedom to make life choices':'freedom to make life choices', 
                         'Overall rank': 'Happiness Rank', 
                         'Score': 'Happiness Score', 
                         'GDP per capita':'Economy (GDP per Capita)', 
                        'Healthy life expectancy': 'Health (Life Expectancy)', 'Dystopia_residual': 'Dystopia Residual'}, inplace = True)
d_2019
```
